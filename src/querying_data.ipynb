```python
# Import the necessary packages. Refer to the Readme.md file for the reason to use the package.
import pandas as pd
import sqlite3
```

```python
# Connecting to database
connection = sqlite3.connect('Freight.db')

# Create a cursor object
curs = connection.cursor()

# Run Update sql query
curs.execute("UPDATE Freights SET Value = '1,800.00' Where Date = '2022-08-12' AND Type='Ocean Freight'")
connection.commit()
```

```python
# Connecting to database
connection = sqlite3.connect('Freight.db')

# Create a cursor object
curs = connection.cursor()

# Run insert sql query
curs.execute("Insert into Freights('Date+Index', 'Type', 'Date', 'Index', 'Value', 'IsOceanFreight') VALUES('2022-08-24_BDI', 'Ocean Freight', '2022-08-24', 'BDI', '1,387.00', 1)")
connection.commit()
```

```python
# Connecting to database
connection = sqlite3.connect('Freight.db')

# Create a cursor object
curs = connection.cursor()

# Run Delete sql query
curs.execute("Delete from Freights Where Date = '2022-08-16'")
connection.commit()
```

```python
# Connecting to database
connection = sqlite3.connect('Freight.db')

# Create a cursor object
curs = connection.cursor()

# Run select sql query
curs.execute('select * from Freights')
 
# Fetch all records
# as list of tuples
records = curs.fetchall()
 
# Display result 
for row in records:
    # show row
    print(row)
```
